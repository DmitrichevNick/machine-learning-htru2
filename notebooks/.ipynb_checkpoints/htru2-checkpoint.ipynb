# Machine Learning

## Dataset: [HTRU2](https://archive.ics.uci.edu/ml/datasets/HTRU2)

### Описание
Датасет содержит кандидатов-пульсаров, собранных в ходе исследования [HTRU](https://sites.google.com/site/htrusouthdeep/home).
Пульсар - разновидность звезды, представляющая значительный научный интерес. Кандидаты должны быть классифицированы в классы "пульсар" и "не пульсар", для дальнейшего более глубоко изучения.

Датасет содержит <strong>17 898</strong> объектов, из которых:
- <strong>16 259</strong> ложных примеров - RFI/шум;
- <strong>639</strong> реальных примеров - настоящие пульсары.

Признаки:

- <strong>mean_IP</strong> - cреднее значение интегрированного профиля;
- <strong>deviation_IP</strong> - стандартное отклонение интегрированного профиля;
- <strong>excess_IP</strong> - лишний эксцесс интегрированного профиля;
- <strong>skewness_IP</strong> - асимметрия интегрированного профиля;
- <strong>mean_DMSNR</strong> - cреднее значение DM-SNR кривой;
- <strong>deviation_DMSNR</strong> - стандартное отклонение DM-SNR кривой;
- <strong>excess_DMSNR</strong> - лишний эксцесс DM-SNR кривой;
- <strong>skewness_DMSNR</strong> - асимметрия DM-SNR кривой;
- <strong>class</strong> - класс объекта: излучение пульсара = 1, помехи и шум = 0.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

### kNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print('Train error:', err_train, '\nTest error: ', err_test)
```

```python
err_test = 1 - knn.score(X_test, y_test)
print('Test error: ', err_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

Ошибок немного.

### Подбор параметров

```python
from sklearn.model_selection import GridSearchCV
nnb = list(range(1, 10)) + list(range(10, 51, 5))
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

Будем работать с тремя соседями.

Посмотрим на ошибки:

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print('Train error:', err_train, '\nTest error: ', err_test)
```

На тренировке ошибка уменьшилась, а на тесте увеличилась.

Посмотрю лучше на график:

```python
kk = list(range(1, 10)) + list(range(10, 20, 5))
err_train = []
err_test = []
err_p = []
for k in kk:
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X_train, y_train)
    err_train.append(np.mean(model.predict(X_train) != y_train))
    err_test.append(np.mean(model.predict(X_test) != y_test))
```

```python
plt.plot(kk, err_train, '.-r', label = 'Train error')
plt.plot(kk, err_test,  '.-b', label = 'Test error')
plt.legend(loc = 2)
```

```python
index_of_min = np.argmin(err_test)
print('Train error:', err_train[index_of_min], '\nTest error: ', err_test[index_of_min])
print(kk[index_of_min])
```

Наименьшая ошибка достигается при k = 7 и равна: 0.01899441340782123

### Random Forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print('Train error:', err_train, '\nTest error: ', err_test)

Err_train, Err_test = err_train, err_test
```

## PCA

#### Сокращаем размерность пространства до двух

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(data)
print("Main components (by rows):\n", pca.components_)
print("Disp aka Eigens:\n", pca.explained_variance_)
print("Deviation:\n", np.sqrt(pca.explained_variance_))
print("Singular values:\n", pca.singular_values_)
```

##### Попытка нарисовать
Рисуем на полученном двумерном пространстве

```python
plt.figure(figsize=(10,10))
for i in range(len(X_pca)):
    plt.scatter(X_pca[i, 0], X_pca[i, 1], c= 'red' if y[i] == 0 else 'blue')
```

На рисунке видно, что **не пульсары** в основном расположены правее **пульсаров**

Случайный лес показал наименьшую ошибку, поэтому применяю его

```python
from sklearn import ensemble
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)

rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print('Train error:', err_train, '\nTest error: ', err_test)
Err_train_pca, Err_test_pca = err_train, err_test
```

```python
print('\tOld\tPCA')
print('train\t', Err_train, '\t', Err_train_pca)
print('test\t', Err_test, '\t', Err_test_pca)
```

Стало хуже :)

#### Построим график зависимости ошибки на тестовой и обучающей выборки в зависимости от количества используемых компонент. Попробуем найти лучшее количество главных компонент

```python
from sklearn.model_selection import train_test_split
end = 9
err_train_list = []
err_test_list = []
for n_comp in range(2, end):
    pca = PCA(n_components=n_comp)
    X_pca = pca.fit_transform(data)
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 42)
    
    rf = ensemble.RandomForestClassifier(n_estimators = 1000)
    rf.fit(X_train, y_train)

    err_train = np.mean(y_train != rf.predict(X_train))
    err_test  = np.mean(y_test  != rf.predict(X_test))
    
    err_train_list.append(err_train)
    err_test_list.append(err_test)
    
plt.plot(range(2, end), err_train_list, label="Train error")
plt.plot(range(2, end), err_test_list, label="Test error")
plt.xlabel('n')
plt.ylabel('error')
plt.legend()
plt.show()
print(min(err_train_list), 2 + err_train_list.index(min(err_train_list)))
min_index = err_test_list.index(min(err_test_list))
Err_train_pca_min, Err_test_pca_min, Pca_comp = err_train_list[min_index], err_test_list[min_index], 2 + min_index
print(err_test_list[min_index], 2 + min_index)
```

Получается, что при использовании семи компонент, ошибка на тесте минимальная. Вообще, в датасете девять признаков; девятый - класс объекта; значит, один из признаков ни на что не влияет?

Значения ошибок с количеством компонент 3, 4, 5, 6 идут почти в одну линию, а разрыв с точкой с семью компонентами довольно большой (0.017)(по крайней мере на графике). Можно пожертвовать точностью 0.017 и взять только три компоненты.

## LDA

```python
from sklearn import discriminant_analysis
lda = discriminant_analysis.LinearDiscriminantAnalysis()
X = data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)  
    
N_train, _ = X_train.shape 
N_test, _ = X_test.shape 
print(N_train, N_test)

lda.fit(X_train, y_train)

err_train = np.mean(y_train != lda.predict(X_train))
err_test = np.mean(y_test != lda.predict(X_test))
print('Train error:', err_train, '\nTest error: ', err_test)
Err_train_lda, Err_test_lda = err_train, err_test
```

```python
print('\ttrain\ttest\tcomp')
print('old\t', Err_train, '\t', Err_test, '\t', '*')
print('pca\t', Err_train_pca, '\t', Err_test_pca, '\t', 2)
print('pca_min\t', Err_train_pca_min, '\t', Err_test_pca_min, '\t', Pca_comp)
print('lda\t', Err_train_lda, '\t', Err_test_lda, '\t', '*')
```

Всё привело к тому, что лучших результатов можно добиться при использовании PCA с семью компонентами и последующим применении RF.
